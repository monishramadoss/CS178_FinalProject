```python
file_count = len(all_files)
```

```python
punctuation = string.punctuation.replace("'", "")
stopset = set(stopwords.words('english'))
```

```python
def cleanWordbank(words : 'list of str'):
    #convert to lowercase
    lower = [word.lower() for word in words]
    
    #remove punctuation from tokens
    punc_filter = str.maketrans('', '', punctuation)
    stripped = [word.translate(punc_filter) for word in lower]
    
    #remove remaining alphanumerics
    words = [word for word in stripped if word.isalpha()]
    
    #filter out basic stopwords
    final_words = [word for word in words if word not in stopset]
    
    return final_words
```

```python
cleaned_wb = cleanWordbank(word_bank)
cleaned_wb2 = cleanWordbank(word_bank2)

word_bank = [str(word) for word in reuters.words()]
FULL = cleanWordbank(word_bank)
FULL_set = set(FULL)

print(len(FULL))
```

```python
freqMap = defaultdict(int)

i = 1
for file in all_files:
    print("FILE")
    rawWords = [str(word) for word in reuters.words(file)]
    cleaned_wb = cleanWordbank(rawWords)
    
    for word in cleaned_wb:
        freqMap[word] += 1
        
    i += 1
    print(round(i / file_count, 2), " out of 1")
freqTuples = list(freqMap.items())
```

```python
# sorting based on frequency
sorted_freq = freqTuples.sort(key= lambda x: x[1], reverse=True)

i = 0
while( i < 50):
    print(freqTuples[i])
    i += 1
```
